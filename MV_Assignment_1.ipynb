```python
import numpy as np
import pandas as pd
import cv2 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
from matplotlib import pyplot as plt
urls = ["https://image.shutterstock.com/image-photo/cannes-france-may-21-leonardo-260nw-1433831474.jpg",
        "https://media1.popsugar-assets.com/files/thumbor/-boRsUHr2Zu22VZE5tETup1bbKQ/0x78:2306x2384/fit-in/2048xorig/filters:format_auto-!!-:strip_icc-!!-/2019/06/19/950/n/1922153/1e5067d35d0aadb3e2ecf9.02908622_/i/Best-Celebrity-Makeup-Artists.jpg",
        "https://media1.popsugar-assets.com/files/thumbor/mqMIr5bPqk2fOEKNkA6jIdFzZQY/548x13:3728x3193/fit-in/2048xorig/filters:format_auto-!!-:strip_icc-!!-/2019/12/03/897/n/1922398/06edd8065de6c673c496a1.76579684_/i/Hottest-Celebrities-Snapchat.jpg",
        "https://cdn.cliqueinc.com/posts/280333/celebrity-new-haircut-280333-1619199407294-main.700x0c.jpg",
        "https://www.lolwhy.com/wp-content/uploads/2020/07/Miley-Cyrus-celebrity-face-closeup.jpg",
        "https://stylecaster.com/wp-content/uploads/2020/12/Drake.jpg?w=503",
        "https://www.gannett-cdn.com/presto/2020/03/25/USAT/eff607b4-10fc-44c5-a05e-1fb6474302c2-GTY_1178240117.JPG?crop=2741,1542,x0,y0&width=660&height=370&format=pjpg&auto=webp",
        "https://static.more.com/wp-content/uploads/2014/07/25214432/SEI_74430577.jpg",
        "https://www.realself.com/news/wp-content/uploads/2021/01/RS_Jennifer-Aniston.jpg",
        "https://messika.cdn-tech.io/media/ultranoir/messikablog/celebrity/image//b/e/beyonce-super-bowl-messika-vignette.jpg"] 
for url in urls:
  img = io.imread(url)
  img1=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  image=cv2.resize(img1,(300,300))
  cv2_imshow(image)
  
          
  blurImg = cv2.blur(image,(20,20)) 
  Imgblur=cv2.resize(blurImg,(300,300))
  cv2_imshow(Imgblur)
  edges = cv2.Canny(image,100,200)
  plt.subplot(122),plt.imshow(edges,cmap = 'gray')
  plt.title('Original Image'), plt.xticks([]), plt.yticks([])
  plt.subplot(122),plt.imshow(edges,cmap = 'gray')
  plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
  plt.show()
  down_width = 50
  down_height = 400
  down_points = (down_width, down_height)
  resized_down = cv2.resize(image, down_points, interpolation= cv2.INTER_LINEAR)
  
  up_width = 40
  up_height = 200
  up_points = (up_width, up_height)
  resized_up = cv2.resize(image, up_points, interpolation= cv2.INTER_LINEAR)
  
  cv2_imshow(resized_down)
  cv2_imshow(resized_up)

  
  
cv2.waitKey(0)
cv2.destroyAllWindows()
```
